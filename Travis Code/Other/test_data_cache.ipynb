```python
import os
import numpy as np
import time
import cv2
import multiprocessing
```

```python
def save_sample(path, arr_view):

    arr_view_index = 0

    sample_full_path = path
    this_example_name = sample_full_path.split(os.sep)[-1]
    print(path)

    for angle in [2, 7]:
      for theta in range(12):
        if angle == 7 and theta == 3:
          continue
        this_image_path = os.path.join(sample_full_path, 'angle_' + "{:02d}".format(angle + 1) + '_theta_' + "{:02d}".format(theta + 1) + '_' + this_example_name + '.jpg')
        image = cv2.imread(this_image_path)
        rgb_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        chan_first = rgb_array.transpose([2, 0, 1])
        #tensor = torch.from_numpy(chan_first)
        arr_view[arr_view_index, ...] = chan_first
        arr_view_index +=1
```

```python
dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'

def fill_arr(this_c, class_arr):
    print(this_c)
    this_count = 0
    full_path = os.path.join(dir_path, this_c, 'train')
    list_dir = os.listdir(full_path)
    for idx, sample in enumerate(list_dir):
      sample_path = os.path.join(full_path, sample)
      save_sample(sample_path, class_arr[idx, ...])
    print('done', this_c)
```

```python
#https://stackoverflow.com/questions/5442910/how-to-use-multiprocessing-pool-map-with-multiple-arguments
if __name__ == '__main__':
  dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'

  classes = os.listdir(dir_path)

  count = 0

  class_counts = []

  for this_c in classes:
      this_count = 0
      full_path = os.path.join(dir_path, this_c, 'test')
      list_dir = os.listdir(full_path)
      for sample in list_dir:
        sample_path = os.path.join(full_path, sample)
        sample_files = os.listdir(sample_path)
        count += 1
        this_count += 1
        for file in sample_files:
          ext = file.split('.')[-1]
          if ext == 'npy':
            os.remove(os.path.join(sample_path, file))
            print('r', sample)
      class_counts.append(this_count)

  print(count)
  full_arr = np.zeros([count, 23, 3, 299, 299], dtype=np.uint8)
  start_idx = 0
  class_arr_views = []
  for idx in range(len(class_counts)):
    end_index = start_idx + class_counts[idx]
    class_arr_views.append(full_arr[start_idx:end_index, ...])
    start_idx = end_index
  

  pool_args = zip(classes, class_arr_views)
  p = multiprocessing.Pool()
  entries = p.starmap(fill_arr, pool_args)

  # for idx, view in enumerate(class_arr_views):
  #   fill_arr(classes[idx], view)
  np.save('/content/drive/MyDrive/Cache/x_test_23_views', full_arr)

```
