```python
def save_sample(path):
    sample_list = []

    sample_full_path = path
    this_example_name = sample_full_path.split(os.sep)[-1]

    for angle in [2, 7]:
      for theta in range(12):
        if angle == 7 and theta == 3:
          continue
        this_image_path = os.path.join(sample_full_path, 'angle_' + "{:02d}".format(angle + 1) + '_theta_' + "{:02d}".format(theta + 1) + '_' + this_example_name + '.jpg')
        in_file = open(this_image_path, "rb") # opening for [r]eading as [b]inary
        data = in_file.read() # if you only wanted to read 512 bytes, do .read(512)
        in_file.close()

        sample_list.append(data)
    
    return sample_list
        
```

```python
dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'

def fill_arr(this_c):
    this_class_list = []
    print(this_c)
    this_count = 0
    full_path = os.path.join(dir_path, this_c, 'train')
    list_dir = os.listdir(full_path)
    for idx, sample in enumerate(list_dir):
      sample_path = os.path.join(full_path, sample)
      this_class_list.append(save_sample(sample_path))
    print('done', this_c)
    return this_class_list
```

```python
if __name__ == '__main__':
  dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'

  classes = os.listdir(dir_path)

  p = multiprocessing.Pool()
  entries = p.map(fill_arr, classes)


  #https://stackoverflow.com/questions/4530611/saving-and-loading-objects-and-using-pickle
  import pickle
  filehandler = open("/content/drive/MyDrive/Cache/x_list_23_view.pkl","wb")
  pickle.dump(entries,filehandler)
```
