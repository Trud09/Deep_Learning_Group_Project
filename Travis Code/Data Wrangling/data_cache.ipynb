```python
import os
import numpy as np
import time
import cv2

import torch

```

```python
def get_sample(path):

    sample_full_path = path
    this_example_name = sample_full_path.split(os.sep)[-1]

    sample_list = []

    views = os.listdir(sample_full_path)

    for view in views:
        this_view_path = os.path.join(sample_full_path,  view)
        image = cv2.imread(this_view_path)
        rgb_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        rgb_array = rgb_array
        chan_first = rgb_array.transpose([2, 0, 1])
        #tensor = torch.from_numpy(chan_first)
        sample_list.append(chan_first)
    sample = np.stack(sample_list, axis=0)
    
    
    
    return sample

def cache_dataset(root, train=True):
    classes = os.listdir(root) # take all files in the root directory

    samples = []
    labels = []

    class_names = []

    if train:
        ds_folder_name = 'train'
    else:
        ds_folder_name = 'test'

    for i, this_class in enumerate(classes):
        
        class_path = os.path.join(root, this_class, ds_folder_name)
        
        this_class_samples = os.listdir(class_path)
        for sample in this_class_samples:
            f_path = os.path.join(class_path, sample)
            samples.append(get_sample(f_path))


        this_class_labels = [np.array(i)] * len(this_class_samples)
        labels.extend(this_class_labels)

    x = np.stack(samples, 0)
    y = np.stack(labels, 0)
    return x, y

dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'
x, y = cache_dataset(dir_path)
```

```python
import os
import numpy as np
import time
import cv2
from itertools import repeat

import torch
from tqdm import tqdm

def save_sample(path):

    sample_full_path = path
    this_example_name = sample_full_path.split(os.sep)[-1]

    sample_list = []

    views = os.listdir(sample_full_path)

    if 'data_arr.npy' in views:
      return

    for view in views:
        this_view_path = os.path.join(sample_full_path,  view)
        image = cv2.imread(this_view_path)
        rgb_array = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        rgb_array = rgb_array
        chan_first = rgb_array.transpose([2, 0, 1])
        #tensor = torch.from_numpy(chan_first)
        sample_list.append(chan_first)
    sample = np.stack(sample_list, axis=0)
    
    f_path = os.path.join(sample_full_path,  'data_arr.npy')
    if not os.path.exists(f_path):
      np.save(f_path, sample)
      print('saved', f_path)




def save_a_class(i, this_class, root, ds_folder_name):
    class_path = os.path.join(root, this_class, ds_folder_name)
    print(class_path)
    
    this_class_sample_paths = os.listdir(class_path)
    this_class_samples = []
    for sample in this_class_sample_paths:
        f_path = os.path.join(class_path, sample)
        save_sample(f_path)
        
    print('done', class_path)

if __name__ == '__main__':
  train = True
  print(os.cpu_count())


  dir_path = '/content/drive/MyDrive/ModelNet10_circling_jpg'

  if train:
      ds_folder_name = 'train'
  else:
      ds_folder_name = 'test'

  classes = os.listdir(dir_path) # take all files in the root directory

  samples = []
  labels = []

  class_names = []

  ind = classes.index('sofa')
  save_a_class(ind, 'sofa', dir_path, 'train')
```
